```python
#import pandas library
import pandas as pd
```

```python
#Load the CSV files into separate DataFrames
file1 = pd.read_csv("C:/Users/mneme/Desktop/Hotel.csv")
file2 = pd.read_csv("C:/Users/mneme/Desktop/Restaurant.csv")
file3 = pd.read_csv("C:/Users/mneme/Desktop/Tourist_site.csv")
```

```python
#Count number of rows per dataframe
num_rows_hotel = len(file1)
num_rows_restaurant = len(file2)
num_rows_tourist = len(file3)
```

```python
#Add Type column to categories data if it is from hotel, restaurant or tourist site
file1['Type'] = ['Hotel'] * num_rows_hotel
file2['Type'] = ['Restaurant'] * num_rows_restaurant
file3['Type'] = ['Tourist Site'] * num_rows_tourist
```

```python
print(file1)
print(file2)
print(file3)
```

```python
#Merging data
data_merged = pd.concat([file1, file2, file3], ignore_index=True, sort=False)
```

```python
#Simplifying or Standardizing name of place
data_merged['Location'] = data_merged['Location'].str.replace("Santa Rosa", "Sta Rosa")
data_merged['Location'] = data_merged['Location'].str.replace("Boljoon", "Cebu")
data_merged['Location'] = data_merged['Location'].str.replace("Pasay, Manila", "Pasay")
data_merged['Location'] = data_merged['Location'].str.replace("Makati, Manila", "Makati")
data_merged['Location'] = data_merged['Location'].str.replace("Paranaque, Manila", "Paranaque")
data_merged['Location'] = data_merged['Location'].str.replace("Sta Rosa, Laguna", "Sta Rosa")
```

```python
data_merged.isnull().sum()
```

```python
data_merged['Review_Feedback'] = data_merged['Review_Feedback'].fillna('No Feedback')
data_merged['Category'].fillna(data_merged['Type'], inplace=True)
```

```python
data_merged.isnull().sum()
```

```python
#Group data by location
data_by_location = data_merged.groupby('Location')
```

```python
# Create and save separate DataFrames per location
location_dataframes = {}
```

```python
for location, group_data in data_by_location:
    location_dataframes[location] = group_data
```

```python
# Access individual DataFrames by location
data1 = location_dataframes.get('Pasay')
data2 = location_dataframes.get('Makati')
data3 = location_dataframes.get('Paranaque')
data4 = location_dataframes.get('Loboc')
data5 = location_dataframes.get('San Fernando')
data6 = location_dataframes.get('Butuan')
data7 = location_dataframes.get('Lipa')
data8 = location_dataframes.get('Ozamis')
data9 = location_dataframes.get('Surigao')
data10 = location_dataframes.get('Tuguegarao City')
data11 = location_dataframes.get('Zamboanga')
data12 = location_dataframes.get('Samal')
data13 = location_dataframes.get('Iloilo City')
data14 = location_dataframes.get('Taguig')
data15 = location_dataframes.get('San Mateo')
data16 = location_dataframes.get('Busuanga')
data17 = location_dataframes.get('Nasugbu')
data18 = location_dataframes.get('Cabanatuan')
data19 = location_dataframes.get('Manila')
data20 = location_dataframes.get('San Juan')
data21 = location_dataframes.get('Sta Rosa')
data22 = location_dataframes.get('Boracay')
data23 = location_dataframes.get('Puerto Princesa')
data24 = location_dataframes.get('Coron')
data25 = location_dataframes.get('El Nido')
data26 = location_dataframes.get('Bohol')
data27 = location_dataframes.get('Banaue')
data28 = location_dataframes.get('Cebu')
data29 = location_dataframes.get('Albay')
data30 = location_dataframes.get('Siquijor')
data31 = location_dataframes.get('Batangas')
data32 = location_dataframes.get('Vigan')
```

```python
# Saving it on CSV file per location
for location, group_data in data_by_location:
    # Create a DataFrame for the current location
    location_df = pd.DataFrame(group_data)
    
    # Define the CSV file name based on the location
    file_name = f"{location}.csv"
    
    # Save the DataFrame to a CSV file
    location_df.to_csv(file_name, index=False)
    
    print(f"Data for '{location}' saved to '{file_name}'")
```
